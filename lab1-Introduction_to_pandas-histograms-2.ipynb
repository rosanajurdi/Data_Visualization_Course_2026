### Explanation of Pandas Functions

**What each function reveals about the data:**

1. **`head()`** - Shows the first 5 (or specified number of) rows, giving us a quick peek at the data structure, column names, and sample values.

2. **`dtypes`** - Displays the data type of each column (e.g., int64, float64, object). This helps us understand what kind of data we're working with (numeric vs categorical).

3. **`shape`** - Returns the dimensions of the DataFrame as (rows, columns). This tells us how much data we have.

4. **`describe()`** - Provides summary statistics for numeric columns: count, mean, standard deviation, min, max, and quartiles (25%, 50%, 75%). This helps understand the distribution and spread of numerical data.

5. **`info()`** - Gives a concise summary including the index dtype, column dtypes, non-null counts, and memory usage. It helps identify missing values and data types at a glance.

6. **`isnull().sum()`** - Counts the number of missing (null) values in each column. This is crucial for data quality assessment before analysis.

## Step 3 – Histograms with Seaborn

### Histogram of price

```python
# Histogram of price using sns.histplot()
plt.figure(figsize=(10, 6))
sns.histplot(diamonds_df['price'], kde=False)
plt.title('Histogram of Diamond Price', fontsize=14)
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

### Histogram of carat

```python
# Histogram of carat using sns.histplot()
plt.figure(figsize=(10, 6))
sns.histplot(diamonds_df['carat'], kde=False)
plt.title('Histogram of Diamond Carat', fontsize=14)
plt.xlabel('Carat', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

### Histogram of price with different bin sizes

```python
# Histogram of price with different bin sizes - 3 side by side
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# bins=10
sns.histplot(diamonds_df['price'], bins=10, kde=False, ax=axes[0], color='steelblue')
axes[0].set_title('Price Histogram (bins=10)', fontsize=12)
axes[0].set_xlabel('Price ($)')
axes[0].set_ylabel('Frequency')

# bins=50
sns.histplot(diamonds_df['price'], bins=50, kde=False, ax=axes[1], color='steelblue')
axes[1].set_title('Price Histogram (bins=50)', fontsize=12)
axes[1].set_xlabel('Price ($)')
axes[1].set_ylabel('Frequency')

# bins=100
sns.histplot(diamonds_df['price'], bins=100, kde=False, ax=axes[2], color='steelblue')
axes[2].set_title('Price Histogram (bins=100)', fontsize=12)
axes[2].set_xlabel('Price ($)')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**Comment on how bin size affects interpretation:**

- **Small number of bins (e.g., 10)**: Shows a very general, smoothed view of the distribution. May hide important details and nuances in the data.

- **Medium number of bins (e.g., 50)**: Provides a good balance between detail and readability. Often reveals the general shape and any peaks in the distribution.

- **Large number of bins (e.g., 100)**: Shows fine-grained details and can reveal subtle patterns, gaps, or outliers. However, can also introduce noise and make the overall pattern harder to see.

Choosing the right bin size is important for accurate data interpretation.

### Histogram of depth

### Histogram of price with distplot (deprecated but included for compatibility)

```python
# sns.distplot() of price with kde=True (normal scale)
# Note: distplot is deprecated but kept for compatibility
plt.figure(figsize=(10, 6))
sns.distplot(diamonds_df['price'], kde=True, bins=50)
plt.title('Distribution Plot of Diamond Price (Normal Scale)', fontsize=14)
plt.xlabel('Price ($)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()
```

### Histogram of log-transformed price with distplot

```python
# sns.distplot() of log(price) with kde=True (log scale)
plt.figure(figsize=(10, 6))
sns.distplot(np.log(diamonds_df['price']), kde=True, bins=50)
plt.title('Distribution Plot of Log-Transformed Price', fontsize=14)
plt.xlabel('Log(Price)', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.show()
```

### Using histplot with KDE as alternative to deprecated distplot

```python
# Alternative using sns.histplot with kde=True (recommended)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Price with KDE
sns.histplot(diamonds_df['price'], kde=True, ax=axes[0], color='steelblue')
axes[0].set_title('Price Distribution with KDE', fontsize=12)
axes[0].set_xlabel('Price ($)')
axes[0].set_ylabel('Frequency')

# Log-transformed price with KDE
sns.histplot(np.log(diamonds_df['price']), kde=True, ax=axes[1], color='coral')
axes[1].set_title('Log-Transformed Price Distribution with KDE', fontsize=12)
axes[1].set_xlabel('Log(Price)')
axes[1].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

### Analysis Questions

**1. Why apply log transformation to 'price'?**

Log transformation is applied to 'price' because the price distribution is typically right-skewed (long tail towards higher prices). The log transformation helps:
- Make the distribution more symmetric (closer to normal)
- Reduce the impact of extreme values/outliers
- Stabilize variance (homoscedasticity)
- Reveal hidden patterns that are not visible in the original scale

**2. What does the log-transformed histogram reveal?**

The log-transformed histogram reveals a more symmetric, bell-shaped distribution. It shows that the data is actually multi-modal (has multiple peaks), which suggests there may be distinct price clusters or categories in the data (possibly related to different diamond qualities or sizes).

**3. What are the approximate peaks in the log distribution and what do they suggest?**

The log-transformed price distribution typically shows 2-3 distinct peaks:
- A major peak around log(price) ≈ 7-8 (roughly $1,000-$3,000)
- A secondary peak at higher values
- These peaks suggest distinct price segments/market categories for diamonds

**4. What are the 3 main characteristics of a distribution identifiable from a histogram?**

1. **Shape**: The overall form - whether it's symmetric, skewed left/right, uniform, or multi-modal
2. **Center**: The central tendency - where the bulk of the data is concentrated (visible through peaks)
3. **Spread**: The variability or dispersion - how spread out the data is (width of the distribution)

**5. How does bin size choice influence histogram interpretation?**

- Too few bins: Can obscure important patterns and details
- Too many bins: Can introduce noise and make patterns hard to see
- Optimal bin size: Reveals the true underlying distribution shape without overfitting or underfitting
- Rule of thumb: Start with Sturges' rule or try multiple bin sizes to find the best balance

### Plotly vs Seaborn Comparison for Histograms

**Interactivity:**
- **Plotly**: Provides fully interactive histograms with hover tooltips, zoom, pan, and the ability to export as HTML or PNG. Users can click legend items to toggle visibility.
- **Seaborn**: Creates static matplotlib figures. No built-in interactivity (though can be made interactive with matplotlib widgets).

**Syntax:**
- **Plotly**: More concise syntax with `px.histogram()`. Easy to create complex visualizations with less code. The `nbins` parameter directly controls the approximate number of bins.
- **Seaborn**: Uses `sns.histplot()` with `bins` parameter for exact bin count. Integrates well with matplotlib's figure-level and axis-level functions.

**Ease of Use:**
- **Plotly**: Very easy for quick exploratory data analysis. The figures are self-contained and interactive by default. Great for web applications and dashboards.
- **Seaborn**: Better integrated with the Python scientific stack (pandas, matplotlib). Easier to customize for publication-quality static figures. More control over fine-grained visual details.

**Best Use Cases:**
- **Plotly**: When sharing analysis online, creating dashboards, or when user interactivity is needed.
- **Seaborn**: When creating figures for papers, reports, or when combining with other matplotlib-based visualizations.

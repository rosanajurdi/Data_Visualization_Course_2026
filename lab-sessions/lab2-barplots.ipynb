##   Introduction aux Diagrammes à Barres

Les diagrammes à barres sont un outil de visualisation fondamental utilisé pour afficher des données catégorielles avec des barres rectangulaires. La longueur de chaque barre est proportionnelle à la valeur qu'elle représente. Les diagrammes à barres sont excellents pour comparer des quantités entre différentes catégories.

**Quand utiliser les diagrammes à barres :**

* Comparaison de données catégorielles.
* Affichage de la fréquence ou du nombre de différentes catégories.
* Visualisation des tendances entre les catégories.

**Terminologie de base :**

* Barres : Les colonnes rectangulaires représentant les valeurs des données.
* Axes : Les lignes horizontales (axe des x) et verticales (axe des y) qui définissent le graphique.
* Étiquettes : Le texte qui décrit les catégories et les valeurs.
* Titre : La description générale du graphique.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt
```

## Chargement et Inspection des Données

```python
data = pd.read_csv('https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/diamonds.csv')
```

```python
data.dtypes
```

```python
data.head(10)
```

## Visualisation sur diagramme à barres

### Matplotlib

```python

# Nombre de diamants par catégorie de 'cut'
cut_counts = data['cut'].value_counts()

plt.bar(cut_counts.index, cut_counts.values)
plt.xlabel("Qualité")
plt.ylabel("Nombre de diamants")
plt.title("Nombre de diamants par qualité")
plt.show()

# syntaxe déclarative un peu répétitive
# Très personnalisable, mais plus longue
# Ca reste facile d'utilisation
```

### Seaborn

```python
fig = sns.catplot(
    data=data,
    x='cut',
    kind='count',
    order=data['cut'].value_counts().index
)

fig.set_axis_labels("Qualité", "Nombre de diamants")
fig.set(title="Nombre de diamants par qualité")

fig
# haut niveau, calcul des effectifs simplifié
# Style par défaut plus propre/joli que Matplotlib
# Personnalisation fine un peu moins flexible
```

### Plotly

```python

cut_counts = data['cut'].value_counts().reset_index()
cut_counts.columns = ['cut', 'count']

fig = px.bar(cut_counts, x='cut', y='count',
             labels={'cut': 'Qualité', 'count': 'Nombre de diamants'},
             title="Nombre de diamants par qualité")
fig.show()

# Graphiques interactifs natifs (info-bulles, zoom, export)
# Syntaxe concise bien qu'un peu plus longue que Seaborn
# Plus lourd que Matplotlib/Seaborn pour du statique mais esthétiquement plus moderne et joli que les autres
```

### Altair

```python
cut_counts_agg = data['cut'].value_counts().reset_index()
cut_counts_agg.columns = ['cut', 'count']

fig = alt.Chart(cut_counts_agg).mark_bar().encode(
    x=alt.X('cut:N', title='Qualité', sort='-y'),
    y=alt.Y('count:Q', title='Nombre de diamants')
).properties(
    title="Nombre de diamants par qualité",
    width=500, height=350
)
fig

# Syntaxe déclarative : on décrit ce qu'on veut, pas comment le faire
# Code compact mais on a moins l'habitude de cette syntaxe donc un peu moins intuitive au début
# Datasets > 5000 lignes nécessitent une pré-agrégation
```

## Analyse Supplémentaire : Prix moyen par qualité

### Matplotlib

```python
mean_price = data.groupby('cut')['price'].mean().reset_index()
mean_price.columns = ['cut', 'mean_price']

plt.bar(mean_price['cut'], mean_price['mean_price'])
plt.xlabel("Qualité")
plt.ylabel("Prix moyen")
plt.title("Prix moyen par qualité")
plt.show()
```

### Seaborn

```python
fig = sns.catplot(
    data=mean_price,
    x='cut',
    y='mean_price',
    kind='bar'
)

fig.set_axis_labels("Qualité", "Prix moyen")
fig.set(title="Prix moyen par qualité")
fig
```

### Plotly

```python

fig = px.bar(mean_price, x='cut', y='mean_price',
              labels={'cut': 'Qualité', 'mean_price': 'Prix moyen'},
              title="Prix moyen par qualité")
fig.show()
```

### Altair

```python

fig = alt.Chart(mean_price).mark_bar().encode(
    x=alt.X('cut:N', title='Qualité', sort='-y'),
    y=alt.Y('mean_price:Q', title='Prix moyen ($)')
).properties(
    title="Prix moyen par qualité",
    width=500, height=350
)
fig
```

##   Introduction aux Diagrammes à Barres

Les diagrammes à barres sont un outil de visualisation fondamental utilisé pour afficher des données catégorielles avec des barres rectangulaires. La longueur de chaque barre est proportionnelle à la valeur qu'elle représente. Les diagrammes à barres sont excellents pour comparer des quantités entre différentes catégories.

**Quand utiliser les diagrammes à barres :**

* Comparaison de données catégorielles.
* Affichage de la fréquence ou du nombre de différentes catégories.
* Visualisation des tendances entre les catégories.

**Terminologie de base :**

* Barres : Les colonnes rectangulaires représentant les valeurs des données.
* Axes : Les lignes horizontales (axe des x) et verticales (axe des y) qui définissent le graphique.
* Étiquettes : Le texte qui décrit les catégories et les valeurs.
* Titre : La description générale du graphique.

```python
# Importation des bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt
```

## 1. Chargement et Inspection des Données

```python
# Chargement des données depuis le fichier local (déjà téléchargé dans le Lab 1)
diamonds_df = pd.read_csv('diamonds.csv')

# Afficher les 5 premières lignes
print("=== 5 premières lignes ===")
diamonds_df.head()
```

```python
# Noms des colonnes et types de données
print("=== Colonnes et types de données ===")
print(diamonds_df.dtypes)
print(f"\nNombre de lignes: {len(diamonds_df)}")
print(f"Nombre de colonnes: {len(diamonds_df.columns)}")
```

```python
# Afficher les 10 premières lignes
print("=== 10 premières lignes ===")
diamonds_df.head(10)
```

```python
# Compter le nombre de diamants par catégorie de 'cut'
cut_counts = diamonds_df['cut'].value_counts()
print("=== Nombre de diamants par catégorie de coupe (cut) ===")
print(cut_counts)
```

<hr style="border: 2px solid black;">

## 2. Visualisation par Diagramme à Barres

### 2.1 Matplotlib

```python
# Diagramme à barres avec Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))

# Créer le diagramme à barres
bars = ax.bar(cut_counts.index, cut_counts.values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])

# Personnalisation
ax.set_xlabel('Qualité de coupe (Cut)', fontsize=12)
ax.set_ylabel('Nombre de diamants', fontsize=12)
ax.set_title('Nombre de diamants par catégorie de coupe - Matplotlib', fontsize=14, fontweight='bold')

# Ajouter les valeurs sur les barres
for bar, value in zip(bars, cut_counts.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 200, 
            f'{value:,}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

**Commentaire sur Matplotlib :**
- **Syntaxe** : Matplotlib utilise une approche orientée objet (fig, ax) qui offre un contrôle total sur chaque élément du graphique.
- **Personnalisation** : Très flexible, permet de personnaliser chaque détail (couleurs, polices, annotations).
- **Facilité d'utilisation** : Code plus verbeux que d'autres bibliothèques, nécessite plus de lignes pour obtenir un résultat esthétique.
- **Forces** : Contrôle précis, large documentation, base de nombreuses autres bibliothèques.
- **Faiblesses** : Syntaxe parfois complexe, pas d'interactivité native.

### 2.2 Seaborn

```python
# Diagramme à barres avec Seaborn
plt.figure(figsize=(10, 6))

# Utiliser countplot pour compter automatiquement les occurrences
ax = sns.countplot(data=diamonds_df, x='cut', 
                   order=['Ideal', 'Premium', 'Very Good', 'Good', 'Fair'],
                   palette='viridis')

# Personnalisation
plt.xlabel('Qualité de coupe (Cut)', fontsize=12)
plt.ylabel('Nombre de diamants', fontsize=12)
plt.title('Nombre de diamants par catégorie de coupe - Seaborn', fontsize=14, fontweight='bold')

# Ajouter les valeurs sur les barres
for container in ax.containers:
    ax.bar_label(container, fmt='%d', fontsize=10)

plt.tight_layout()
plt.show()
```

**Commentaire sur Seaborn :**
- **Syntaxe** : Plus concise que Matplotlib. `countplot` compte automatiquement les occurrences.
- **Style par défaut** : Graphiques plus esthétiques dès le départ avec des palettes de couleurs intégrées.
- **Différences avec Matplotlib** : Seaborn simplifie le code en gérant automatiquement le comptage des données catégorielles.
- **Forces** : Code plus court, intégration native avec pandas, styles modernes.
- **Faiblesses** : Moins de contrôle fin que Matplotlib, construit sur Matplotlib donc dépendant.

### 2.3 Plotly

```python
# Diagramme à barres interactif avec Plotly
# Préparer les données
cut_order = ['Ideal', 'Premium', 'Very Good', 'Good', 'Fair']
cut_data = diamonds_df['cut'].value_counts().reindex(cut_order)

fig = px.bar(x=cut_data.index, y=cut_data.values,
             labels={'x': 'Qualité de coupe (Cut)', 'y': 'Nombre de diamants'},
             title='Nombre de diamants par catégorie de coupe - Plotly',
             color=cut_data.index,
             color_discrete_sequence=px.colors.qualitative.Set2,
             text=cut_data.values)

fig.update_traces(textposition='outside')
fig.update_layout(
    showlegend=False,
    template='plotly_white',
    font=dict(size=12)
)
fig.show()
```

**Commentaire sur Plotly :**
- **Interactivité** : Principal avantage ! Info-bulles au survol, zoom, pan, export d'image intégrés.
- **Syntaxe** : API de haut niveau (px) très intuitive, similaire à Seaborn.
- **Différences** : Graphiques interactifs par défaut vs statiques pour Matplotlib/Seaborn.
- **Forces** : Interactivité native, rendu moderne, export HTML possible, excellent pour les dashboards.
- **Faiblesses** : Fichiers plus lourds, nécessite JavaScript pour l'affichage.

### 2.4 Altair

```python
# Diagramme à barres avec Altair (syntaxe déclarative)
# Altair a une limite de 5000 lignes par défaut, on utilise les données agrégées
cut_df = diamonds_df['cut'].value_counts().reset_index()
cut_df.columns = ['cut', 'count']

chart = alt.Chart(cut_df).mark_bar().encode(
    x=alt.X('cut:N', 
            sort=['Ideal', 'Premium', 'Very Good', 'Good', 'Fair'],
            title='Qualité de coupe (Cut)'),
    y=alt.Y('count:Q', title='Nombre de diamants'),
    color=alt.Color('cut:N', legend=None, scale=alt.Scale(scheme='category10')),
    tooltip=['cut', 'count']
).properties(
    title='Nombre de diamants par catégorie de coupe - Altair',
    width=600,
    height=400
)

# Ajouter les étiquettes de texte
text = chart.mark_text(
    align='center',
    baseline='bottom',
    dy=-5
).encode(
    text='count:Q'
)

(chart + text)
```

**Commentaire sur Altair :**
- **Syntaxe déclarative** : On décrit CE QUE l'on veut visualiser, pas COMMENT le dessiner. Approche basée sur la grammaire des graphiques (Grammar of Graphics).
- **Encodages** : x, y, color, size sont des "canaux" auxquels on associe des variables.
- **Comparaison** : Syntaxe plus abstraite que Matplotlib mais très puissante pour des visualisations complexes.
- **Forces** : Syntaxe élégante, interactivité, exports multiples (PNG, SVG, HTML).
- **Faiblesses** : Limite de 5000 lignes par défaut (contournable), courbe d'apprentissage pour la syntaxe.

## 3. Analyse Supplémentaire

### 3.1 Diagramme à barres par couleur (color)

```python
# Diagramme à barres par couleur avec Plotly
color_counts = diamonds_df['color'].value_counts().sort_index()

fig = px.bar(x=color_counts.index, y=color_counts.values,
             labels={'x': 'Couleur du diamant', 'y': 'Nombre de diamants'},
             title='Nombre de diamants par couleur - Plotly',
             color=color_counts.index,
             color_discrete_sequence=px.colors.sequential.Viridis,
             text=color_counts.values)

fig.update_traces(textposition='outside')
fig.update_layout(showlegend=False, template='plotly_white')
fig.show()
```

### 3.2 Prix moyen par qualité de coupe (cut)

```python
# Prix moyen par qualité de coupe - Comparaison des 4 bibliothèques
avg_price_by_cut = diamonds_df.groupby('cut')['price'].mean().reindex(
    ['Ideal', 'Premium', 'Very Good', 'Good', 'Fair']
)

# Créer une figure avec 2x2 subplots
fig, axes = plt.subplots(2, 2, figsize=(14, 10))

# 1. Matplotlib
axes[0, 0].bar(avg_price_by_cut.index, avg_price_by_cut.values, color='steelblue')
axes[0, 0].set_xlabel('Qualité de coupe')
axes[0, 0].set_ylabel('Prix moyen ($)')
axes[0, 0].set_title('Matplotlib', fontweight='bold')
axes[0, 0].tick_params(axis='x', rotation=45)

# 2. Seaborn
sns.barplot(x=avg_price_by_cut.index, y=avg_price_by_cut.values, ax=axes[0, 1], palette='Blues_d')
axes[0, 1].set_xlabel('Qualité de coupe')
axes[0, 1].set_ylabel('Prix moyen ($)')
axes[0, 1].set_title('Seaborn', fontweight='bold')
axes[0, 1].tick_params(axis='x', rotation=45)

# Masquer les deux derniers subplots (on utilise Plotly et Altair séparément)
axes[1, 0].axis('off')
axes[1, 1].axis('off')
axes[1, 0].text(0.5, 0.5, 'Voir graphique Plotly ci-dessous', ha='center', va='center', fontsize=12)
axes[1, 1].text(0.5, 0.5, 'Voir graphique Altair ci-dessous', ha='center', va='center', fontsize=12)

plt.suptitle('Prix moyen des diamants par qualité de coupe', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Prix moyen par qualité de coupe - Plotly
avg_price_df = avg_price_by_cut.reset_index()
avg_price_df.columns = ['cut', 'avg_price']

fig = px.bar(avg_price_df, x='cut', y='avg_price',
             labels={'cut': 'Qualité de coupe', 'avg_price': 'Prix moyen ($)'},
             title='Prix moyen des diamants par qualité de coupe - Plotly',
             color='cut',
             text=avg_price_df['avg_price'].round(0).astype(int))

fig.update_traces(textposition='outside')
fig.update_layout(showlegend=False, template='plotly_white')
fig.show()
```

```python
# Prix moyen par qualité de coupe - Altair
chart = alt.Chart(avg_price_df).mark_bar().encode(
    x=alt.X('cut:N', 
            sort=['Ideal', 'Premium', 'Very Good', 'Good', 'Fair'],
            title='Qualité de coupe'),
    y=alt.Y('avg_price:Q', title='Prix moyen ($)'),
    color=alt.Color('cut:N', legend=None),
    tooltip=['cut', alt.Tooltip('avg_price:Q', format='.0f')]
).properties(
    title='Prix moyen des diamants par qualité de coupe - Altair',
    width=500,
    height=350
)

text = chart.mark_text(align='center', baseline='bottom', dy=-5).encode(
    text=alt.Text('avg_price:Q', format='.0f')
)

(chart + text)
```

## 4. Conclusion - Comparaison des bibliothèques

| Critère | Matplotlib | Seaborn | Plotly | Altair |
|---------|------------|---------|--------|--------|
| **Syntaxe** | Verbeuse, orientée objet | Concise, haut niveau | Intuitive (px) | Déclarative |
| **Personnalisation** | Très flexible | Bonne | Excellente | Bonne |
| **Interactivité** | Non native | Non native | Native | Native |
| **Courbe d'apprentissage** | Moyenne | Facile | Facile | Moyenne |
| **Cas d'utilisation** | Publications, contrôle fin | Exploration rapide | Dashboards, web | Analyses déclaratives |
| **Style par défaut** | Basique | Esthétique | Moderne | Moderne |

**Recommandations :**
- **Matplotlib** : Pour un contrôle total et des publications scientifiques
- **Seaborn** : Pour l'exploration rapide de données avec un style agréable
- **Plotly** : Pour des visualisations interactives et des dashboards
- **Altair** : Pour une approche déclarative et des visualisations complexes

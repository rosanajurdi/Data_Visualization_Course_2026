##   Introduction aux Diagrammes à Barres

Les diagrammes à barres sont un outil de visualisation fondamental utilisé pour afficher des données catégorielles avec des barres rectangulaires. La longueur de chaque barre est proportionnelle à la valeur qu'elle représente. Les diagrammes à barres sont excellents pour comparer des quantités entre différentes catégories.

**Quand utiliser les diagrammes à barres :**

* Comparaison de données catégorielles.
* Affichage de la fréquence ou du nombre de différentes catégories.
* Visualisation des tendances entre les catégories.

**Terminologie de base :**

* Barres : Les colonnes rectangulaires représentant les valeurs des données.
* Axes : Les lignes horizontales (axe des x) et verticales (axe des y) qui définissent le graphique.
* Étiquettes : Le texte qui décrit les catégories et les valeurs.
* Titre : La description générale du graphique.

## 2. Visualisation par Diagramme à Barres

### 2.1 Matplotlib

**Commentaire :** Matplotlib offre un contrôle total sur chaque élément du graphique (axes, couleurs, tailles, etc.), mais la syntaxe est plus verbeuse et nécessite plus de lignes de code pour obtenir un résultat esthétique. C'est la bibliothèque de base sur laquelle les autres s'appuient.

```python
# =====================================================
# 2.1 Diagramme à barres avec Matplotlib
# =====================================================
import matplotlib.pyplot as plt

# Compter le nombre de diamants par catégorie de cut
cut_counts = diamonds_df['cut'].value_counts()

# Créer le diagramme à barres
fig, ax = plt.subplots(figsize=(8, 5))
ax.bar(cut_counts.index, cut_counts.values, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd'])
ax.set_xlabel('Qualité de la Taille (Cut)')
ax.set_ylabel('Nombre de Diamants')
ax.set_title('Nombre de Diamants par Catégorie de Cut - Matplotlib')
plt.tight_layout()
plt.show()
```

### 2.2 Seaborn

**Commentaire :** Seaborn simplifie beaucoup la création de graphiques statistiques. Avec `sns.countplot()`, on obtient directement le décompte sans calcul préalable. Le style par défaut est plus esthétique que Matplotlib. Seaborn gère automatiquement les palettes de couleurs et les thèmes, ce qui réduit le code nécessaire. En revanche, la personnalisation fine reste moins flexible qu'avec Matplotlib pur.

```python
# =====================================================
# 2.2 Diagramme à barres avec Seaborn
# =====================================================
import seaborn as sns

fig, ax = plt.subplots(figsize=(8, 5))
sns.countplot(data=diamonds_df, x='cut', 
              order=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal'],
              palette='viridis', ax=ax)
ax.set_xlabel('Qualité de la Taille (Cut)')
ax.set_ylabel('Nombre de Diamants')
ax.set_title('Nombre de Diamants par Catégorie de Cut - Seaborn')
plt.tight_layout()
plt.show()
```

### 2.3 Plotly

**Commentaire :** Plotly se distingue par son **interactivité native** : survol (info-bulles avec les valeurs exactes), zoom, pan, et export en image, le tout sans code supplémentaire. La syntaxe de `plotly.express` est très concise et proche de celle de Seaborn (une seule ligne pour le graphique). Contrairement à Matplotlib et Seaborn qui produisent des images statiques, Plotly génère du HTML/JS interactif, ce qui le rend idéal pour des dashboards ou des présentations. En revanche, les graphiques Plotly sont plus lourds en mémoire et moins adaptés à l'impression papier.

```python
# =====================================================
# 2.3 Diagramme à barres interactif avec Plotly
# =====================================================
import plotly.express as px

# Compter le nombre de diamants par cut
cut_counts_df = diamonds_df['cut'].value_counts().reset_index()
cut_counts_df.columns = ['cut', 'count']

# Ordonner les catégories
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
cut_counts_df['cut'] = pd.Categorical(cut_counts_df['cut'], categories=cut_order, ordered=True)
cut_counts_df = cut_counts_df.sort_values('cut')

fig = px.bar(cut_counts_df, x='cut', y='count',
             color='cut',
             labels={'cut': 'Qualité de la Taille (Cut)', 'count': 'Nombre de Diamants'},
             title='Nombre de Diamants par Catégorie de Cut - Plotly')
fig.update_layout(xaxis_title='Qualité de la Taille (Cut)',
                  yaxis_title='Nombre de Diamants')
fig.show()
```

### 2.4 Altair

**Commentaire :** Altair utilise une **syntaxe déclarative** basée sur la grammaire des graphiques (Vega-Lite). Au lieu de décrire *comment* dessiner le graphique étape par étape (comme Matplotlib), on décrit *ce que* l'on veut visualiser : les données, les encodages (axes, couleurs) et le type de marque (barres). Le code est très lisible et concis. Altair gère nativement l'agrégation (`count()`) directement dans l'encodage, sans avoir besoin de pré-calculer les valeurs. Comme Plotly, les graphiques sont interactifs par défaut (info-bulles). En revanche, Altair est limité à ~5000 lignes par défaut (il faut activer explicitement les grands jeux de données).

```python
# =====================================================
# 2.4 Diagramme à barres avec Altair
# =====================================================
import altair as alt

# Altair a une limite par défaut de 5000 lignes, on la désactive
alt.data_transformers.enable('vegafusion')

chart = alt.Chart(diamonds_df).mark_bar().encode(
    x=alt.X('cut:N', title='Qualité de la Taille (Cut)',
            sort=['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']),
    y=alt.Y('count():Q', title='Nombre de Diamants'),
    color=alt.Color('cut:N', legend=alt.Legend(title='Cut')),
    tooltip=['cut:N', 'count():Q']
).properties(
    title='Nombre de Diamants par Catégorie de Cut - Altair',
    width=500,
    height=350
)

chart
```

## 3. Analyse Supplémentaire (Bonus)

### Prix moyen par catégorie de Cut avec Plotly

**Commentaire :** Pour cette analyse bonus, j'utilise Plotly car il permet de visualiser les agrégations de manière interactive. L'agrégation (calcul de la moyenne) est effectuée en amont avec `pandas` via `groupby().mean()`, puis le résultat est passé directement à Plotly pour la visualisation. Plotly n'effectue pas l'agrégation lui-même (contrairement à Altair qui peut le faire dans l'encodage avec `mean()`), mais la combinaison pandas + plotly.express reste très fluide et lisible. L'interactivité permet de survoler chaque barre pour voir le prix moyen exact.

```python
# =====================================================
# 3. Analyse Supplémentaire : Prix moyen par Cut
# =====================================================

# Agrégation avec pandas
avg_price_by_cut = diamonds_df.groupby('cut')['price'].mean().reset_index()
avg_price_by_cut.columns = ['cut', 'avg_price']

# Ordonner les catégories
cut_order = ['Fair', 'Good', 'Very Good', 'Premium', 'Ideal']
avg_price_by_cut['cut'] = pd.Categorical(avg_price_by_cut['cut'], categories=cut_order, ordered=True)
avg_price_by_cut = avg_price_by_cut.sort_values('cut')

# Visualisation avec Plotly
fig = px.bar(avg_price_by_cut, x='cut', y='avg_price',
             color='cut',
             labels={'cut': 'Qualité de la Taille (Cut)', 'avg_price': 'Prix Moyen (USD)'},
             title='Prix Moyen des Diamants par Catégorie de Cut - Plotly',
             text='avg_price')
fig.update_traces(texttemplate='$%{text:.0f}', textposition='outside')
fig.update_layout(xaxis_title='Qualité de la Taille (Cut)',
                  yaxis_title='Prix Moyen (USD)',
                  showlegend=False)
fig.show()
```

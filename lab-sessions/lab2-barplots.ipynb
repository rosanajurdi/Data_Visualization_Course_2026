##   Introduction aux Diagrammes à Barres

Les diagrammes à barres sont un outil de visualisation fondamental utilisé pour afficher des données catégorielles avec des barres rectangulaires. La longueur de chaque barre est proportionnelle à la valeur qu'elle représente. Les diagrammes à barres sont excellents pour comparer des quantités entre différentes catégories.

**Quand utiliser les diagrammes à barres :**

* Comparaison de données catégorielles.
* Affichage de la fréquence ou du nombre de différentes catégories.
* Visualisation des tendances entre les catégories.

**Terminologie de base :**

* Barres : Les colonnes rectangulaires représentant les valeurs des données.
* Axes : Les lignes horizontales (axe des x) et verticales (axe des y) qui définissent le graphique.
* Étiquettes : Le texte qui décrit les catégories et les valeurs.
* Titre : La description générale du graphique.

```python
## Step 1 Chargement et Inspection des Données

## Import des Libraries
import pandas as pd
```

```python
# Charger les données depuis l'URL
url = "https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/diamonds.csv"
diamonds = pd.read_csv(url)

# Affichez les 5 premières lignes de l'ensemble de données pour comprendre sa structure.
print("5 premières lignes :")
print(diamonds.head())

# Imprimez les noms des colonnes et leurs types de données.
print("\nColonnes et types :")
print(diamonds.dtypes)

# Affichez les 10 premières lignes de l'ensemble de données.
print("\n10 premières lignes :")
print(diamonds.head(10))
```

```python
## Step 2: Visualisation par Diagramme à Barres

## 2.1 : Matplotlib

# Comptage du nombre de diamants par 'cut'
cut_counts = diamonds['cut'].value_counts()

# Création du diagramme à barres
plt.figure(figsize=(8,5))
plt.bar(cut_counts.index, cut_counts.values, color='skyblue')
plt.xlabel("Cut")
plt.ylabel("Nombre de diamants")
plt.title("Nombre de diamants par catégorie de cut (Matplotlib)")
plt.show()
```

```python
## 2.2 : Seaborn
import seaborn as sns

# Création du diagramme à barres
plt.figure(figsize=(8,5))
sns.countplot(data=diamonds, x='cut', palette='pastel')
plt.xlabel("Cut")
plt.ylabel("Nombre de diamants")
plt.title("Nombre de diamants par catégorie de cut (Seaborn)")
plt.show()
```

```python
## 2.3 Plotly

import plotly.express as px

# Diagramme à barres interactif
fig = px.bar(diamonds['cut'].value_counts().reset_index(),
             x='cut', y='count',
             labels={'cut':'Cut', 'count':'Nombre de diamants'},
             title="Nombre de diamants par catégorie de cut (Plotly)")
fig.show()
```

```python
## 2.4 : Altair

import altair as alt

# Préparer les données
cut_data = diamonds['cut'].value_counts().reset_index()
cut_data.columns = ['cut','count']

# Diagramme à barres
chart = alt.Chart(cut_data).mark_bar(color='orange').encode(
    x='cut',
    y='count'
).properties(
    title='Nombre de diamants par catégorie de cut (Altair)',
    width=500,
    height=300
)

chart.show()
```

```python
## Step 3: Analyse Supplémentaire (Facultatif)

# Moyenne du prix par cut
avg_price = diamonds.groupby('cut')['price'].mean().reset_index()

# Seaborn exemple
plt.figure(figsize=(8,5))
sns.barplot(data=avg_price, x='cut', y='price', palette='muted')
plt.xlabel("Cut")
plt.ylabel("Prix moyen")
plt.title("Prix moyen des diamants par cut (Seaborn)")
plt.show()
```

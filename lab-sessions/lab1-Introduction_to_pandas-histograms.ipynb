<font color="#008000"> **Sélectionner des colonnes d'un DataFrame. **</font>


Pour sélectionner des colonnes spécifiques d'un jeu de données (DataFrame) Pandas, deux méthodes principales sont disponibles : l'opérateur point (`.`) et l'opérateur crochets (`[]`).

**Accès simple :**

Par exemple, pour accéder à la colonne nommée `cut` du DataFrame `diamonds_df`, vous pouvez utiliser :

```python
diamonds_df.cut

# ou 
diamonds_df['cut']
```

Cas où l'opérateur point (.) ne peut pas être utilisé :

L'opérateur point est pratique, mais il présente des limitations dans certains cas :

Noms de colonnes contenant des espaces : Si un nom de colonne inclut des espaces (par exemple, 'prix au carat'), vous devez impérativement utiliser la notation avec les crochets :

```python
diamonds_df['prix au carat']
```

Noms de colonnes entiers : Lorsque le nom d'une colonne est un nombre entier (par exemple, 2023), l'accès via l'opérateur point n'est pas valide. Utilisez les crochets :

```python
diamonds_df[2023]
```

Création de nouvelles colonnes : Pour ajouter une nouvelle colonne au DataFrame, la notation avec les crochets est la méthode standard :

```python
diamonds_df['nouvelle_colonne'] = ...
diamonds_df['une autre colonne'] = ...
```

````python
import plotly.express as pex
import plotly.graph_objects as go
from scipy import stats
import numpy as np


# 1. Histogramme carats

fig = pex.histogram(diamonds_df,
                   x='carat',
                   title='Distribution des carats',
                   labels={'carat': 'Carat', 'count': 'Fréquence'})

fig.update_layout(
    xaxis_title='Carat',
    yaxis_title='Fréquence',
    template='plotly_white',
    showlegend=False
)

fig.show()

# 2. Histogramme avec 50 bins

fig = pex.histogram(diamonds_df,
                   x='carat',
                   nbins=50,
                   title='Distribution des carats (50 bins)',
                   labels={'carat': 'Carat'})

fig.update_layout(
    xaxis_title='Carat',
    yaxis_title='Fréquence',
    template='plotly_white',
    showlegend=False
)

fig.show()


# 3. Histogramme avec courbe KDE

# Calculer la KDE
density = stats.gaussian_kde(diamonds_df['carat'])
xs = np.linspace(diamonds_df['carat'].min(), diamonds_df['carat'].max(), 200)
ys = density(xs)

# Créer la figure
fig = go.Figure()

# Ajouter l'histogramme
fig.add_trace(go.Histogram(
    x=diamonds_df['carat'],
    name='Histogramme',
    opacity=0.7,
    histnorm='probability density'
))

# Ajouter la courbe KDE
fig.add_trace(go.Scatter(
    x=xs,
    y=ys,
    mode='lines',
    name='KDE',
    line=dict(color='red', width=2)
))

fig.update_layout(
    title='Distribution des carats avec KDE',
    xaxis_title='Carat',
    yaxis_title='Densité',
    template='plotly_white',
    showlegend=True
)

fig.show()


# 4. Histogramme sans KDE, 100 bins

fig = pex.histogram(diamonds_df,
                   x='carat',
                   nbins=100,
                   title='Distribution des carats (100 bins, sans KDE)',
                   labels={'carat': 'Carat'})

fig.update_layout(
    xaxis_title='Carat',
    yaxis_title='Fréquence',
    template='plotly_white',
    showlegend=False
)

fig.show()

# 5. Transformation logarithmique du prix avec KDE

# Calculer le log des prix
log_price = np.log(diamonds_df['price'])

# Calculer la KDE pour les prix transformés
density_log = stats.gaussian_kde(log_price)
xs_log = np.linspace(log_price.min(), log_price.max(), 200)
ys_log = density_log(xs_log)

# Créer la figure
fig = go.Figure()

# Ajouter l'histogramme
fig.add_trace(go.Histogram(
    x=log_price,
    name='Histogramme',
    opacity=0.7,
    histnorm='probability density'
))

# Ajouter la courbe KDE
fig.add_trace(go.Scatter(
    x=xs_log,
    y=ys_log,
    mode='lines',
    name='KDE',
    line=dict(color='red', width=2)
))

fig.update_layout(
    title='Distribution du logarithme du prix avec KDE',
    xaxis_title='log(Prix)',
    yaxis_title='Densité',
    template='plotly_white',
    showlegend=True
)
````

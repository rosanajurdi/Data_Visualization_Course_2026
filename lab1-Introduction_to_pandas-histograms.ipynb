# Pandas Data Exploration and Histogram Visualization

This notebook demonstrates data exploration using Pandas and histogram visualizations using Seaborn and Plotly on the diamonds dataset.


## Setup

```python
import pandas as pd
import seaborn as sns
import numpy as np
import plotly.express as px

diamonds_df = sns.load_dataset('diamonds')
```

## Part 1 – Data Exploration with Pandas

```python
# Task 1: Display the first 5 rows using head()
diamonds_df.head()
```

```python
# Task 2: Display the first 10 rows using head(10)
diamonds_df.head(10)
```

```python
# Task 3: Get the shape of the DataFrame (rows x columns)
diamonds_df.shape
```

```python
# Task 4: Use describe() to get summary statistics for numeric columns
diamonds_df.describe()
```

```python
# Task 5: Use describe(include=object) to summarize categorical columns (cut, color, clarity)
diamonds_df.describe(include=object)
```

```python
# Task 6: Add a new column price_per_carat = price / carat
diamonds_df['price_per_carat'] = diamonds_df['price'] / diamonds_df['carat']
diamonds_df.head()
```

```python
# Task 7: Use np.where() to create a column price_per_carat_is_high (1 if price_per_carat > 3500, else 0)
diamonds_df['price_per_carat_is_high'] = np.where(diamonds_df['price_per_carat'] > 3500, 1, 0)
diamonds_df.head()
```

## Part 2 – Histograms with Pandas/Seaborn

```python
# Task 8: Plot a histogram of carat using diamonds_df.hist(column='carat')
diamonds_df.hist(column='carat')
```

```python
# Task 9: Plot the same histogram with bins=50 and observe how more bins affect granularity
diamonds_df.hist(column='carat', bins=50)
```

```python
# Task 10: Plot using sns.histplot(diamonds_df['carat'])
# Note: distplot is deprecated, use histplot or displot instead
sns.histplot(diamonds_df['carat'])
```

```python
# Task 11: Plot with KDE disabled: sns.histplot(diamonds_df['carat'], kde=False)
sns.histplot(diamonds_df['carat'], kde=False)
```

```python
# Task 12: Plot with bins=100 and KDE disabled
sns.histplot(diamonds_df['carat'], kde=False, bins=100)
```

```python
# Task 13: Apply a log transformation and plot: sns.histplot(np.log(diamonds_df['price']), kde=True)
# Log transformation is useful for skewed data like price because it:
# - Makes right-skewed distributions more symmetric
# - Helps identify patterns that aren't visible at original scale
# - Stabilizes variance
sns.histplot(np.log(diamonds_df['price']), kde=True)
```

## Part 3 – Repeat visualizations with Plotly

```python
# Task 14: Histogram of carat: px.histogram(diamonds_df, x='carat')
# This visualization shows the distribution of diamond carat weights
fig = px.histogram(diamonds_df, x='carat', title='Histogram of Diamond Carat')
fig.show()
```

```python
# Task 15: Histogram of carat with nbins=50
# Using more bins provides more granularity in the distribution
fig = px.histogram(diamonds_df, x='carat', nbins=50, title='Histogram of Diamond Carat (50 bins)')
fig.show()
```

```python
# Task 16: Histogram of carat with nbins=100
# Even more bins reveal finer details in the distribution
fig = px.histogram(diamonds_df, x='carat', nbins=100, title='Histogram of Diamond Carat (100 bins)')
fig.show()
```

```python
# Task 17: Histogram of log-transformed price: px.histogram(diamonds_df, x=np.log(diamonds_df['price']), nbins=50, labels={'x': 'log(price)'})
# This shows the price distribution on a log scale, revealing distinct price clusters
fig = px.histogram(diamonds_df, x=np.log(diamonds_df['price']), nbins=50, 
                    labels={'x': 'log(price)'}, title='Log-Transformed Price Distribution')
fig.show()
```

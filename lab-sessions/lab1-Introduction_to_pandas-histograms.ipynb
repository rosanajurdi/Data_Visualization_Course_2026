<font color="#008000"> **Sélectionner des colonnes d'un DataFrame. **</font>


Pour sélectionner des colonnes spécifiques d'un jeu de données (DataFrame) Pandas, deux méthodes principales sont disponibles : l'opérateur point (`.`) et l'opérateur crochets (`[]`).

**Accès simple :**

Par exemple, pour accéder à la colonne nommée `cut` du DataFrame `diamonds_df`, vous pouvez utiliser :

```python
diamonds_df.cut

# ou 
diamonds_df['cut']
```

Cas où l'opérateur point (.) ne peut pas être utilisé :

L'opérateur point est pratique, mais il présente des limitations dans certains cas :

Noms de colonnes contenant des espaces : Si un nom de colonne inclut des espaces (par exemple, 'prix au carat'), vous devez impérativement utiliser la notation avec les crochets :

```python
diamonds_df['prix au carat']
```

Noms de colonnes entiers : Lorsque le nom d'une colonne est un nombre entier (par exemple, 2023), l'accès via l'opérateur point n'est pas valide. Utilisez les crochets :

```python
diamonds_df[2023]
```

Création de nouvelles colonnes : Pour ajouter une nouvelle colonne au DataFrame, la notation avec les crochets est la méthode standard :

```python
diamonds_df['nouvelle_colonne'] = ...
diamonds_df['une autre colonne'] = ...
```



Maintenant, nous allons examiner l'ajout conditionnel de colonnes. Essayons d'ajouter une colonne basée sur la valeur de `price_per_carat`. Disons que toute valeur supérieure à 3500 sera considérée comme "élevée" (codée par 1) et toute valeur inférieure à 3500 sera considérée comme "faible" (codée par 0).

````python
#Operations on Dataframes: 

import numpy as np
diamonds_df['price_per_carat'] = diamonds_df['price']/diamonds_df['carat']

# Ajout colonne binaire : 1 si price_per_carat > 3500, sinon 0
diamonds_df['price_level'] = np.where(diamonds_df['price_per_carat'] > 3500, 1, 0)
diamonds_df.head()
````

**Q3 : Effet de l'augmentation du nombre de bins**

Avec plus de bins (50 vs ~10 par défaut), on voit apparaître des pics nets sur les valeurs rondes de carats (0.3, 0.5, 1.0, 1.5, 2.0). Avec peu de bins, ces pics sont noyés dans des barres trop larges. Par contre, trop de bins (genre 200+) donne un graphique bruité et illisible.

````python
# Histogramme Seaborn avec courbe KDE
sns.histplot(diamonds_df['carat'], kde=True)
plt.show()
````

````python
# Histogramme Seaborn sans KDE
sns.histplot(diamonds_df['carat'], kde=False)
plt.show()
````

````python
# Histogramme Seaborn avec 100 bins, sans KDE
sns.histplot(diamonds_df['carat'], kde=False, bins=100)
plt.show()
````

**Q5 : Deux différences clés pandas `.hist()` vs Seaborn `histplot()`**

1. **KDE** — Seaborn a l'option `kde=True` pour tracer directement la courbe de densité par-dessus l'histogramme. Avec pandas `.hist()` c'est pas possible sans rajouter du code matplotlib à la main.
2. **Style** — Le rendu par défaut de Seaborn est beaucoup plus propre (couleurs, grille, espacement). Pandas produit un truc assez brut visuellement, pensé pour l'exploration rapide.

**Q6 : Qu'est-ce que le KDE ?**

KDE = Kernel Density Estimation. C'est une courbe lissée qui approxime la distribution des données. Chaque point du dataset génère une petite bosse gaussienne, et on les additionne toutes pour obtenir la courbe finale. L'avantage par rapport à l'histogramme : le résultat ne dépend pas du choix des bins, donc on voit mieux la "vraie forme" de la distribution.

````python
# Transformation logarithmique sur le prix + KDE
import numpy as np
sns.histplot(np.log(diamonds_df['price']), kde=True)
plt.show()
````

<hr style="border: 2px solid black;">


**Pourquoi appliquer une transformation logarithmique sur 'price' ?**

Le prix des diamants est très asymétrique à droite : beaucoup de diamants pas chers et quelques-uns très chers. L'histogramme brut est écrasé à gauche. Le log "étale" les petites valeurs et compresse les grandes, ce qui donne une distribution plus lisible.

**Que révèle l'histogramme en log ?**

On voit apparaître une distribution quasi bimodale :
- Un premier pic autour de log(price) ≈ 6.8, soit environ ~$900 — ce sont les diamants d'entrée de gamme (petits carats, tailles courantes)
- Un deuxième pic plus large entre 8.5 et 9.0, soit environ ~$4500-$8000 — diamants plus gros/de meilleure qualité

**Pics et interprétation**

Ces deux groupes correspondent probablement à deux segments du marché : les diamants "standards" accessibles et les diamants haut de gamme. Le creux entre les deux pics (~7.5, soit ~$1800) marque la frontière entre ces segments.

**Analyse et discussion**

**3 caractéristiques d'une distribution identifiables sur un histogramme :**
- **Forme** — symétrique ou asymétrique ? unimodale ou bimodale ? Ici les carats sont clairement asymétriques à droite (queue vers les gros carats)
- **Centre** — où se concentrent les données. Pour les carats, la majorité est entre 0.2 et 1.5 ct.
- **Dispersion** — l'étendue des valeurs. Les carats vont de ~0.2 à ~5 ct, mais 99% sont en dessous de 3 ct.

**Influence du choix des bins :**
- Peu de bins (ex: 10) → on voit la tendance globale mais on rate les pics sur 0.5, 1.0, etc.
- Beaucoup de bins (ex: 50-100) → on voit ces pics mais le graphique commence à être chargé
- Trop de bins (ex: 500) → c'est du bruit, chaque barre a très peu d'observations

## Repeat the visualizations but with Plotly

**Bibliothèques utilisées :**
[Pandas](https://pandas.pydata.org/docs/) | [Plotly](https://plotly.com/python-api-reference/)

````python
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
import pandas as pd
import seaborn as sns

# Charger les données
diamonds_df = sns.load_dataset('diamonds')

# === Pré-agréger les données pour réduire la taille des outputs Plotly ===

# 1. Histogramme simple de carat (30 bins) - pré-agrégé
counts1, edges1 = np.histogram(diamonds_df['carat'], bins=30)
centers1 = (edges1[:-1] + edges1[1:]) / 2
fig1 = px.bar(x=centers1, y=counts1,
              title='Distribution des Carats (Histogramme simple)',
              labels={'x': 'Carat', 'y': 'Fréquence'})
fig1.update_layout(bargap=0.05)
fig1.show()

# 2. Histogramme avec 50 bins - pré-agrégé
counts2, edges2 = np.histogram(diamonds_df['carat'], bins=50)
centers2 = (edges2[:-1] + edges2[1:]) / 2
fig2 = px.bar(x=centers2, y=counts2,
              title='Distribution des Carats (50 bins)',
              labels={'x': 'Carat', 'y': 'Fréquence'})
fig2.update_layout(bargap=0.05)
fig2.show()

# 3. Histogramme avec violin marginal - pré-agrégé + violin séparé
counts3, edges3 = np.histogram(diamonds_df['carat'], bins=30)
centers3 = (edges3[:-1] + edges3[1:]) / 2
fig3 = go.Figure()
fig3.add_trace(go.Bar(x=centers3, y=counts3, name='Histogramme'))
fig3.update_layout(title='Distribution des Carats (30 bins)',
                   xaxis_title='Carat', yaxis_title='Fréquence', bargap=0.05)
fig3.show()

# 4. Histogramme 100 bins - pré-agrégé
counts4, edges4 = np.histogram(diamonds_df['carat'], bins=100)
centers4 = (edges4[:-1] + edges4[1:]) / 2
fig4 = px.bar(x=centers4, y=counts4,
              title='Distribution des Carats (100 bins - haute résolution)',
              labels={'x': 'Carat', 'y': 'Fréquence'})
fig4.update_layout(bargap=0.05)
fig4.show()

# 5. Transformation logarithmique du prix - pré-agrégé
log_price = np.log(diamonds_df['price'])
counts5, edges5 = np.histogram(log_price, bins=50)
centers5 = (edges5[:-1] + edges5[1:]) / 2
fig5 = px.bar(x=centers5, y=counts5,
              title='Distribution des Prix en Échelle Logarithmique',
              labels={'x': 'Log(Prix)', 'y': 'Fréquence'})
fig5.update_layout(bargap=0.05)
fig5.show()

# 6. Heatmap 2D carat vs price - pré-agrégé
heatmap, xedges, yedges = np.histogram2d(diamonds_df['carat'], diamonds_df['price'], bins=30)
fig6 = px.imshow(heatmap.T, origin='lower',
                 x=np.round((xedges[:-1] + xedges[1:]) / 2, 2),
                 y=np.round((yedges[:-1] + yedges[1:]) / 2, 0).astype(int),
                 title='Relation entre Carat et Prix',
                 labels={'x': 'Carat', 'y': 'Prix (USD)', 'color': 'Count'},
                 aspect='auto')
fig6.show()
````

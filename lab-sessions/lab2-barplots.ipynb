##   Introduction aux Diagrammes à Barres

Les diagrammes à barres sont un outil de visualisation fondamental utilisé pour afficher des données catégorielles avec des barres rectangulaires. La longueur de chaque barre est proportionnelle à la valeur qu'elle représente. Les diagrammes à barres sont excellents pour comparer des quantités entre différentes catégories.

**Quand utiliser les diagrammes à barres :**

* Comparaison de données catégorielles.
* Affichage de la fréquence ou du nombre de différentes catégories.
* Visualisation des tendances entre les catégories.

**Terminologie de base :**

* Barres : Les colonnes rectangulaires représentant les valeurs des données.
* Axes : Les lignes horizontales (axe des x) et verticales (axe des y) qui définissent le graphique.
* Étiquettes : Le texte qui décrit les catégories et les valeurs.
* Titre : La description générale du graphique.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt
import warnings
warnings.filterwarnings('ignore')
```

## 1. Chargement et Inspection des Données

```python
# Charger les données
url = "https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/diamonds.csv"
df = pd.read_csv(url)

# Afficher les 5 premières lignes
print("5 premières lignes du dataset:")
print(df.head())

# Noms des colonnes et types de données
print("\nNoms des colonnes et types de données:")
print(df.dtypes)

# Afficher les 10 premières lignes
print("\n10 premières lignes du dataset:")
print(df.head(10))

print(f"\nDimensions du dataset: {df.shape}")
print(f"\nStatistiques descriptives:")
print(df.describe())
```

## 2. Visualisation par Diagramme à Barres

### 2.1. Matplotlib

```python
# Préparer les données pour le diagramme à barres
diamonds_by_cut = df['cut'].value_counts().sort_index()
print("Nombre de diamants par 'cut':")
print(diamonds_by_cut)

# Créer le diagramme à barres avec Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
diamonds_by_cut.plot(kind='bar', ax=ax, color='steelblue', edgecolor='black')
ax.set_xlabel('Qualité de Coupe (Cut)', fontsize=12, fontweight='bold')
ax.set_ylabel('Nombre de Diamants', fontsize=12, fontweight='bold')
ax.set_title('Distribution des Diamants par Qualité de Coupe\n(Matplotlib)', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.3)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 2.2. Seaborn

```python
# Créer le diagramme à barres avec Seaborn
diamonds_by_cut_df = df['cut'].value_counts().reset_index()
diamonds_by_cut_df.columns = ['cut', 'count']

fig, ax = plt.subplots(figsize=(10, 6))
sns.barplot(data=diamonds_by_cut_df, x='cut', y='count', palette='Set2', ax=ax, edgecolor='black')
ax.set_xlabel('Qualité de Coupe (Cut)', fontsize=12, fontweight='bold')
ax.set_ylabel('Nombre de Diamants', fontsize=12, fontweight='bold')
ax.set_title('Distribution des Diamants par Qualité de Coupe\n(Seaborn)', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

### 2.3. Plotly

```python
# Créer un diagramme à barres interactif avec Plotly
diamonds_by_cut_df = df['cut'].value_counts().reset_index()
diamonds_by_cut_df.columns = ['cut', 'count']

fig = px.bar(diamonds_by_cut_df, x='cut', y='count',
             labels={'cut': 'Qualité de Coupe', 'count': 'Nombre de Diamants'},
             title='Distribution des Diamants par Qualité de Coupe (Plotly)',
             color='count',
             color_continuous_scale='Viridis')
fig.update_layout(
    plot_bgcolor='rgba(0,0,0,0)',
    hovermode='x unified',
    font=dict(size=12),
    height=600,
    width=1000
)
fig.show()
```

### 2.4. Altair

```python
# Créer un diagramme à barres avec Altair
diamonds_by_cut_df = df['cut'].value_counts().reset_index()
diamonds_by_cut_df.columns = ['cut', 'count']

chart = alt.Chart(diamonds_by_cut_df).mark_bar(color='steelblue', opacity=0.7).encode(
    x=alt.X('cut:N', title='Qualité de Coupe'),
    y=alt.Y('count:Q', title='Nombre de Diamants'),
    tooltip=['cut', 'count']
).properties(
    width=600,
    height=400,
    title='Distribution des Diamants par Qualité de Coupe (Altair)'
).interactive()

chart.show()
```

## 3. Analyse Supplémentaire

### 3.1. Distribution par Couleur (Color)

```python
# Analyse par couleur
print("Distribution des diamants par couleur (Color):")
color_counts = df['color'].value_counts()
print(color_counts)

# Visualiser avec Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
df['color'].value_counts().sort_index().plot(kind='bar', ax=ax, color='coral', edgecolor='black')
ax.set_xlabel('Couleur (Color)', fontsize=12, fontweight='bold')
ax.set_ylabel('Nombre de Diamants', fontsize=12, fontweight='bold')
ax.set_title('Distribution des Diamants par Couleur', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.3)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 3.2. Prix Moyen par Qualité de Coupe (Cut)

```python
# Prix moyen par qualité de coupe
print("Prix moyen par qualité de coupe (Cut):")
avg_price_by_cut = df.groupby('cut')['price'].mean().sort_values(ascending=False)
print(avg_price_by_cut)

# Visualiser avec Matplotlib
fig, ax = plt.subplots(figsize=(10, 6))
avg_price_by_cut.plot(kind='bar', ax=ax, color='lightgreen', edgecolor='black')
ax.set_xlabel('Qualité de Coupe (Cut)', fontsize=12, fontweight='bold')
ax.set_ylabel('Prix Moyen ($)', fontsize=12, fontweight='bold')
ax.set_title('Prix Moyen des Diamants par Qualité de Coupe', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.3)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

### 3.3. Prix Moyen par Couleur (Color)

```python
# Prix moyen par couleur
print("Prix moyen par couleur (Color):")
avg_price_by_color = df.groupby('color')['price'].mean().sort_values(ascending=False)
print(avg_price_by_color)

# Visualiser avec Seaborn
fig, ax = plt.subplots(figsize=(10, 6))
avg_price_df = avg_price_by_color.reset_index()
avg_price_df.columns = ['color', 'avg_price']
sns.barplot(data=avg_price_df, x='color', y='avg_price', palette='YlOrRd', ax=ax, edgecolor='black')
ax.set_xlabel('Couleur (Color)', fontsize=12, fontweight='bold')
ax.set_ylabel('Prix Moyen ($)', fontsize=12, fontweight='bold')
ax.set_title('Prix Moyen des Diamants par Couleur', fontsize=14, fontweight='bold')
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
```

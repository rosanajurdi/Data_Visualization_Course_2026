##   Introduction aux Diagrammes à Barres

Les diagrammes à barres sont un outil de visualisation fondamental utilisé pour afficher des données catégorielles avec des barres rectangulaires. La longueur de chaque barre est proportionnelle à la valeur qu'elle représente. Les diagrammes à barres sont excellents pour comparer des quantités entre différentes catégories.

**Quand utiliser les diagrammes à barres :**

* Comparaison de données catégorielles.
* Affichage de la fréquence ou du nombre de différentes catégories.
* Visualisation des tendances entre les catégories.

**Terminologie de base :**

* Barres : Les colonnes rectangulaires représentant les valeurs des données.
* Axes : Les lignes horizontales (axe des x) et verticales (axe des y) qui définissent le graphique.
* Étiquettes : Le texte qui décrit les catégories et les valeurs.
* Titre : La description générale du graphique.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt

url = "https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/diamonds.csv"
df = pd.read_csv(url)

print(df.head())
print(df.dtypes)
print(df.head(10))
```

## 2a. Matplotlib — Diagramme à barres (nombre de diamants par coupe)

```python
cut_counts = df['cut'].value_counts()
plt.bar(cut_counts.index, cut_counts.values)
plt.xlabel("Coupe")
plt.ylabel("Nombre de diamants")
plt.title("Diamants par coupe (Matplotlib)")
plt.show()
```

**Matplotlib :** Faut tout écrire à la main, c'est un peu long mais ça marche bien. On contrôle tout mais c'est pas le plus rapide.

## 2b. Seaborn — Diagramme à barres (nombre de diamants par coupe)

```python
sns.countplot(data=df, x='cut')
plt.xlabel("Coupe")
plt.ylabel("Nombre de diamants")
plt.title("Diamants par coupe (Seaborn)")
plt.show()
```

**Seaborn :** Beaucoup plus simple, `countplot` fait le compte tout seul. Le rendu est plus joli par défaut que Matplotlib avec moins de code.

## 2c. Plotly — Diagramme à barres interactif (nombre de diamants par coupe)

```python
cut_counts = df['cut'].value_counts().reset_index()
cut_counts.columns = ['cut', 'count']
fig = px.bar(cut_counts, x='cut', y='count', title="Diamants par coupe (Plotly)",
             labels={'cut': 'Coupe', 'count': 'Nombre de diamants'})
fig.show()
```

**Plotly :** Le graphique est interactif, on peut zoomer et voir les valeurs en passant la souris. C'est le plus pratique pour explorer les données.

## 2d. Altair — Diagramme à barres (nombre de diamants par coupe)

```python
cut_counts = df['cut'].value_counts().reset_index()
cut_counts.columns = ['cut', 'count']

alt.Chart(cut_counts).mark_bar().encode(
    x='cut:N',
    y='count:Q'
).properties(title="Diamants par coupe (Altair)")
```

**Altair :** La syntaxe est différente, on décrit ce qu'on veut afficher plutôt que comment le dessiner. C'est lisible mais faut s'y habituer. Attention, il supporte pas les gros datasets directement.

## 3. Analyse Supplémentaire (Facultatif)

```python
sns.countplot(data=df, x='color')
plt.xlabel("Couleur")
plt.ylabel("Nombre de diamants")
plt.title("Diamants par couleur (Seaborn)")
plt.show()
```

```python
avg_cut = df.groupby('cut')['price'].mean().reset_index()
avg_cut.columns = ['cut', 'avg_price']

fig3 = px.bar(avg_cut, x='cut', y='avg_price',
              title="Prix moyen par coupe (Plotly)",
              labels={'cut': 'Coupe', 'avg_price': 'Prix moyen ($)'})
fig3.show()
```

```python
avg_color = df.groupby('color')['price'].mean().reset_index()
avg_color.columns = ['color', 'avg_price']

alt.Chart(avg_color).mark_bar().encode(
    x='color:N',
    y='avg_price:Q',
    tooltip=['color:N', 'avg_price:Q']
).properties(title="Prix moyen par couleur (Altair)")
```

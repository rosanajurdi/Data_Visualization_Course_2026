##   Introduction aux Diagrammes à Barres

Les diagrammes à barres sont un outil de visualisation fondamental utilisé pour afficher des données catégorielles avec des barres rectangulaires. La longueur de chaque barre est proportionnelle à la valeur qu'elle représente. Les diagrammes à barres sont excellents pour comparer des quantités entre différentes catégories.

**Quand utiliser les diagrammes à barres :**

* Comparaison de données catégorielles.
* Affichage de la fréquence ou du nombre de différentes catégories.
* Visualisation des tendances entre les catégories.

**Terminologie de base :**

* Barres : Les colonnes rectangulaires représentant les valeurs des données.
* Axes : Les lignes horizontales (axe des x) et verticales (axe des y) qui définissent le graphique.
* Étiquettes : Le texte qui décrit les catégories et les valeurs.
* Titre : La description générale du graphique.

```python
# --- IMPORTATIONS ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt
from IPython.display import display

# Désactiver la limite de lignes d'Altair (car on a plus de 5000 diamants)
# alt.data_transformers.disable_max_rows()

# --- 1. CHARGEMENT ET INSPECTION DES DONNÉES ---
url = "https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/diamonds.csv"
diamonds_df = pd.read_csv(url)

print("=== 1. INSPECTION DES DONNÉES ===")
print("\n--- Les 5 premières lignes ---")
display(diamonds_df.head(5))

print("\n--- Noms des colonnes et types de données ---")
print(diamonds_df.dtypes)

print("\n--- Les 10 premières lignes ---")
display(diamonds_df.head(10))


# --- 2. VISUALISATION PAR DIAGRAMME À BARRES ---
print("\n=== 2. DIAGRAMMES À BARRES ET ANALYSE ===")

# Préparation des données pour les graphiques (comptage)
cut_counts = diamonds_df['cut'].value_counts().reset_index()
cut_counts.columns = ['cut', 'count']

# -- A. MATPLOTLIB --
plt.figure(figsize=(8, 5))
plt.bar(cut_counts['cut'], cut_counts['count'], color='skyblue', edgecolor='black')
plt.title("Matplotlib : Nombre de diamants par catégorie de coupe")
plt.xlabel("Qualité de la coupe (Cut)")
plt.ylabel("Nombre de diamants")
plt.show()

print("""
Analyse Matplotlib :
- Syntaxe : Assez bas niveau. Il faut calculer manuellement les occurrences avant de tracer.
- Personnalisation : Très grande, on contrôle chaque détail (couleur, bordure).
- Facilité : Un peu verbeux pour des tâches simples d'exploration.
""")

# -- B. SEABORN --
plt.figure(figsize=(8, 5))
sns.countplot(data=diamonds_df, x='cut', hue='cut', palette='viridis', legend=False)
plt.title("Seaborn : Nombre de diamants par catégorie de coupe")
plt.xlabel("Qualité de la coupe (Cut)")
plt.ylabel("Nombre de diamants")
plt.show()

print("""
Analyse Seaborn :
- Différence avec Matplotlib : Seaborn fait l'agrégation automatiquement (via countplot) !
- Style : Le design par défaut est plus moderne et agréable avec des palettes de couleurs intégrées.
""")

# -- C. PLOTLY --
fig_plotly = px.bar(
    cut_counts, x='cut', y='count', color='cut',
    title="Plotly : Nombre de diamants par catégorie de coupe",
    labels={'cut': 'Coupe', 'count': 'Nombre'},
    template='plotly_white'
)
fig_plotly.show()

print("""
Analyse Plotly :
- Interactivité : C'est son grand point fort (zoom, info-bulles au survol).
- Syntaxe : Très concise (similaire à Seaborn) mais génère un rendu web interactif.
""")

# -- D. ALTAIR --
fig_altair = alt.Chart(cut_counts).mark_bar().encode(
    x=alt.X('cut:N', title='Qualité de la coupe'),
    y=alt.Y('count:Q', title='Nombre de diamants'),
    color='cut:N',
    tooltip=['cut', 'count']
).properties(
    title="Altair : Nombre de diamants par catégorie de coupe",
    width=500, height=300
)
display(fig_altair)

print("""
Analyse Altair :
- Syntaxe : Déclarative. On déclare 'quoi' afficher (axes x et y) plutôt que 'comment' le dessiner.
- Typage : Altair gère intelligemment les types de données (:N pour nominal, :Q pour quantitatif).
""")


# --- 3. ANALYSE SUPPLÉMENTAIRE (FACULTATIF) ---
print("\n=== 3. BONUS : PRIX MOYEN PAR COULEUR (PLOTLY) ===")
# Agrégation : prix moyen par couleur
avg_price_color = diamonds_df.groupby('color')['price'].mean().reset_index()

fig_bonus = px.bar(
    avg_price_color, x='color', y='price', color='price',
    title="Prix moyen des diamants selon leur couleur",
    labels={'color': 'Couleur (D=Meilleur, J=Pire)', 'price': 'Prix moyen ($)'},
    color_continuous_scale='Reds',
    template='plotly_white'
)
fig_bonus.show()
```

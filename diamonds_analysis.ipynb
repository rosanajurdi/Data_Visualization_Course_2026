# Diamond Data Analysis - Bar Charts Visualization

This notebook demonstrates creating bar charts using four popular Python visualization libraries:
- Matplotlib
- Seaborn
- Plotly
- Altair

We'll analyze the diamonds dataset to visualize counts by cut category.

## Task 1: Load and Inspect the Data

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import altair as alt

# Load the diamonds dataset from the URL
url = "https://raw.githubusercontent.com/TrainingByPackt/Interactive-Data-Visualization-with-Python/master/datasets/diamonds.csv"
diamonds = pd.read_csv(url)

# Display the first 5 rows
print("First 5 rows of the dataset:")
display(diamonds.head())
```

```python
# Print column names and data types
print("Column names and data types:")
print(diamonds.dtypes)
print("\n" + "="*50)
print("\nColumn names:", diamonds.columns.tolist())
```

```python
# Display the first 10 rows
print("First 10 rows of the dataset:")
display(diamonds.head(10))
```

## Task 2: Bar Charts for Diamond 'Cut' Counts

Let's visualize the count of diamonds for each 'cut' category using different visualization libraries.

### 1. Matplotlib Bar Chart

Matplotlib is the foundational plotting library in Python. It offers fine-grained control over every aspect of the chart.

```python
# Get count of diamonds per cut category
cut_counts = diamonds['cut'].value_counts()

# Create the bar chart using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(cut_counts.index, cut_counts.values, color='steelblue', edgecolor='black')

# Add labels and title
plt.xlabel('Cut Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Diamond Count by Cut Category (Matplotlib)', fontsize=14, fontweight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add value labels on top of bars
for i, v in enumerate(cut_counts.values):
    plt.text(i, v + 200, str(v), ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

**Comment on Matplotlib:**

Matplotlib provides extensive customization options but requires more verbose code. You need to manually create the figure, set up axes, and add each element individually. While powerful, the syntax can feel low-level compared to other libraries. However, this gives you complete control over every visual aspect of the chart.

### 2. Seaborn Bar Chart

Seaborn is built on top of Matplotlib and provides a higher-level interface with attractive default styles.

```python
# Create the bar chart using Seaborn
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=cut_counts.index, y=cut_counts.values, palette='viridis')

# Add labels and title
plt.xlabel('Cut Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Diamond Count by Cut Category (Seaborn)', fontsize=14, fontweight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add value labels on top of bars
for i, v in enumerate(cut_counts.values):
    ax.text(i, v + 200, str(v), ha='center', fontsize=10)

plt.tight_layout()
plt.show()
```

**Comment on Seaborn vs Matplotlib:**

Seaborn offers a more streamlined syntax with attractive default styling out of the box. The color palettes are visually appealing without additional configuration. It handles aggregation automatically in many cases, reducing code complexity. Seaborn is ideal when you want quick, professional-looking visualizations with minimal customization.

### 3. Plotly Bar Chart

Plotly creates interactive visualizations that can be explored in the notebook or saved as HTML files.

```python
# Create the bar chart using Plotly
fig = px.bar(
    x=cut_counts.index, 
    y=cut_counts.values,
    labels={'x': 'Cut Category', 'y': 'Count'},
    title='Diamond Count by Cut Category (Plotly)'
)

# Update layout for better appearance
fig.update_layout(
    xaxis_title='Cut Category',
    yaxis_title='Count',
    title_font_size=16,
    width=800,
    height=500
)

# Update bar color and add hover text
fig.update_traces(marker_color='steelblue', text=cut_counts.values, textposition='outside')

fig.show()
```

**Comment on Plotly's Interactivity:**

Plotly excels in interactivity - charts automatically include tooltips showing data values on hover, and you can zoom, pan, and save the chart as a PNG. The hover functionality provides instant access to detailed information without additional code. This makes Plotly perfect for exploratory data analysis where you need to dig into specific data points.

### 4. Altair Bar Chart

Altair is a declarative visualization library based on Vega-Lite. It uses a concise JSON-like syntax.

```python
# Prepare data for Altair (needs to be in a DataFrame format)
cut_df = pd.DataFrame({
    'cut': cut_counts.index,
    'count': cut_counts.values
})

# Create the bar chart using Altair
chart = alt.Chart(cut_df).mark_bar(color='steelblue').encode(
    x=alt.X('cut', title='Cut Category', sort='-y'),
    y=alt.Y('count', title='Count'),
    tooltip=['cut', 'count']
).properties(
    title='Diamond Count by Cut Category (Altair)',
    width=600,
    height=400
).configure_title(
    fontSize=16
)

chart
```

**Comment on Altair's Declarative Syntax:**

Altair uses a declarative approach where you specify what to encode (x, y, color) rather than how to draw. This results in concise, readable code. The syntax maps directly to Vega-Lite specifications, making it easy to understand the data transformation pipeline. Altair is excellent for creating reproducible, publication-quality visualizations with minimal code.

## Task 3 (Bonus): Additional Bar Charts

Let's explore additional visualizations for 'color' and 'clarity' categories, and average price analysis.

### Bonus 1: Bar Chart for 'Color' Category

```python
# Get count of diamonds per color category
color_counts = diamonds['color'].value_counts()

# Create the bar chart using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(color_counts.index, color_counts.values, color='coral', edgecolor='black')
plt.xlabel('Color Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Diamond Count by Color Category', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nColor category distribution:")
print(color_counts)
```

### Bonus 2: Bar Chart for 'Clarity' Category

```python
# Get count of diamonds per clarity category
clarity_counts = diamonds['clarity'].value_counts()

# Create the bar chart using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=clarity_counts.index, y=clarity_counts.values, palette='plasma')
plt.xlabel('Clarity Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.title('Diamond Count by Clarity Category', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("\nClarity category distribution:")
print(clarity_counts)
```

### Bonus 3: Average Price by Cut (Matplotlib)

```python
# Calculate average price by cut using groupby
avg_price_cut = diamonds.groupby('cut')['price'].mean().sort_values()

# Create bar chart using Matplotlib
plt.figure(figsize=(10, 6))
plt.bar(avg_price_cut.index, avg_price_cut.values, color='teal', edgecolor='black')
plt.xlabel('Cut Category', fontsize=12)
plt.ylabel('Average Price ($)', fontsize=12)
plt.title('Average Price by Cut Category', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')

# Add value labels
for i, v in enumerate(avg_price_cut.values):
    plt.text(i, v + 50, f'${v:.0f}', ha='center', fontsize=9)

plt.tight_layout()
plt.show()

print("\nAverage price by cut:")
print(avg_price_cut)
```

### Bonus 4: Average Price by Color (Seaborn)

```python
# Calculate average price by color
avg_price_color = diamonds.groupby('color')['price'].mean()

# Create bar chart using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_price_color.index, y=avg_price_color.values, palette='coolwarm')
plt.xlabel('Color Category', fontsize=12)
plt.ylabel('Average Price ($)', fontsize=12)
plt.title('Average Price by Color Category', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()

print("\nAverage price by color:")
print(avg_price_color)
```

### Bonus 5: Average Price by Cut using Plotly (Interactive)

```python
# Calculate average price by cut
avg_price_cut_df = diamonds.groupby('cut')['price'].mean().reset_index()
avg_price_cut_df.columns = ['cut', 'avg_price']

# Create interactive bar chart using Plotly
fig = px.bar(
    avg_price_cut_df, 
    x='cut', 
    y='avg_price',
    labels={'cut': 'Cut Category', 'avg_price': 'Average Price ($)'},
    title='Average Price by Cut Category (Interactive Plotly)',
    color='avg_price',
    color_continuous_scale='Viridis'
)

fig.update_layout(
    xaxis_title='Cut Category',
    yaxis_title='Average Price ($)',
    width=800,
    height=500
)

fig.show()
```

### Bonus 6: Average Price by Cut and Color using Altair

```python
# Calculate average price by cut and color
avg_price_cut_color = diamonds.groupby(['cut', 'color'])['price'].mean().reset_index()
avg_price_cut_color.columns = ['cut', 'color', 'avg_price']

# Create interactive chart using Altair
chart = alt.Chart(avg_price_cut_color).mark_bar().encode(
    x=alt.X('color', title='Color'),
    y=alt.Y('avg_price', title='Average Price ($)'),
    color='color',
    tooltip=['cut', 'color', 'avg_price']
).properties(
    width=600,
    height=400
).facet(
    facet=alt.Facet('cut', title='Cut Category'),
    columns=3
).properties(
    title='Average Price by Cut and Color (Altair Facet)'
)

chart
```

## Summary

This notebook demonstrated creating bar charts using four popular Python visualization libraries:

1. **Matplotlib** - Low-level control, verbose syntax, maximum customization
2. **Seaborn** - Attractive defaults, built-in statistical visualizations, simple API
3. **Plotly** - Interactive charts with tooltips and zoom, web-ready output
4. **Altair** - Declarative syntax, concise code, excellent for data exploration

Each library handles aggregation differently:
- **Matplotlib/Seaborn**: Often require manual aggregation using `groupby()` or `value_counts()`
- **Plotly**: Can handle aggregation automatically with the `histogram` or by passing pre-aggregated data
- **Altair**: Supports automatic aggregation in some mark types, but typically works best with pre-aggregated data

Choose the library based on your needs:
- For quick exploratory analysis: Seaborn or Plotly
- For interactive dashboards: Plotly
- For publication-quality static plots: Matplotlib or Altair
- For reproducible research: Altair (declarative, JSON-based specs)
